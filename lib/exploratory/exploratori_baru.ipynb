```python
# General data analysis/plotting
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt


import matplotlib
import seaborn as sns

df = pd.read_excel("data_clean.xlsx")
df = df[['Diagnosis', 'Tindakan', 'SEX', 'UMUR_TAHUN', 'LAMA_DIRAWAT', 'INACBG', 'HARI']]

df['DiagnosisCAT'] = df['Diagnosis']
df['Diagnosis'] = df['DiagnosisCAT'].astype('category')
df['Diagnosis'] = df['Diagnosis'].cat.reorder_categories(df['DiagnosisCAT'].unique(), ordered=True)
df['Diagnosis'] = df['Diagnosis'].cat.codes

df['TindakanCAT'] = df['Tindakan']
df['Tindakan'] = df['TindakanCAT'].astype('category')
df['Tindakan'] = df['Tindakan'].cat.reorder_categories(df['TindakanCAT'].unique(), ordered=True)
df['Tindakan'] = df['Tindakan'].cat.codes

severity = []
for index, row in df.iterrows():
    if row['INACBG'].split("-")[-1] == "I":
        severity.append(1)
    elif row['INACBG'].split("-")[-1] == "II":
        severity.append(2)
    else:
        severity.append(3)

df['Severity'] = severity

df.dropna(axis=0, inplace=True)
```

```python
severity1 = df[df["Severity"] == 1]
sns.scatterplot(data=severity1, x="UMUR_TAHUN", y="LAMA_DIRAWAT")
plt.title("Persebaran data lama rawat berdasarkan umur tahun pada severity level 1")
```

```python
severity2 = df[df["Severity"] == 2]
sns.scatterplot(data=severity2, x="UMUR_TAHUN", y="LAMA_DIRAWAT")
plt.title("Persebaran data lama rawat berdasarkan umur tahun pada severity level 2")
```

```python
severity3 = df[df["Severity"] == 3]
sns.scatterplot(data=severity3, x="UMUR_TAHUN", y="LAMA_DIRAWAT")
plt.title("Persebaran data lama rawat berdasarkan umur tahun pada severity level 3")
```

```python
data1 = df
sns.set(style="whitegrid")

sns.barplot(data=data1, x="SEX", y="LAMA_DIRAWAT", hue="Severity")
plt.title("Pengaruh gender dengan lama rawat")
plt.xlabel("GENDER")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Contoh data
data = {
    'Category': ['A', 'B', 'C', 'D'],
    'Value_Hue1': [10, 15, 8, 12],
    'Value_Hue2': [5, 8, 6, 10]
}

# Mengubah data menjadi DataFrame Seaborn
df = pd.DataFrame(data)

# Mengatur gaya plot
sns.set(style="whitegrid")

# Membuat barplot dengan dua hue berbeda
plt.figure(figsize=(10, 6))
sns.barplot(x="Category", y="Value_Hue1", data=df, color="blue", label="Hue 1")
sns.barplot(x="Category", y="Value_Hue2", data=df, color="orange", label="Hue 2")

# Menambahkan legenda
plt.legend()

plt.title("Barplot dengan Dua Hue Berbeda pada Satu Barplot")
plt.xlabel("Category")
plt.ylabel("Value")
plt.show()
```
